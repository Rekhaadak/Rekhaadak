```python
#4.Write a python program to scrap book name, author name, genre and book review of any 5 books from‘www.bookpage.com
from bs4 import BeautifulSoup
import requests

page=requests.get("https://bookpage.com/reviews")
page
soup=BeautifulSoup(page.content)
soup
bookname=soup.find_all("h4",class_="italic")
Authorname=soup.find_all("p",class_="sans bold")
genre=soup.find_all("p",class_="genre-links hidden-phone")
bookreview=soup.find_all("p", class_="excerpt")

b_name=[]
a_name=[]
b_genre=[]
b_review=[]

srno=[]
for i in range(1,6):
    srno.append(i)
    
for i in bookname:
    for j in i.find_all("a"):
        b_name.append(j.text.replace("\n",""))

for i in Authorname:
    a_name.append(i.text.replace("\n",""))

for i in genre:
    b_genre.append(i.text.replace("\n",""))

for i in bookreview:
    b_review.append(i.text.replace("\n",""))
    

bookinfo=pd.DataFrame({})
bookinfo['Sr.no']=srno
bookinfo['Book Name']=b_name[:5]
bookinfo['Author Name']=a_name[:5]
bookinfo['Book Genre']=b_genre[:5]
bookinfo['Book review']=b_review[:5]
bookinfo
```

```python
#9.		Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, company name, CTC, and apply date.
from bs4 import BeautifulSoup
import requests
page=requests.get("https://internshala.com/fresher-jobs")
page
soup=BeautifulSoup(page.content)
title=soup.find_all("div",class_="heading_4_5 profile")

j_title=[]
for i in title:
    for j in i.find_all("a"):
        j_title.append(i.text.replace("\n",""))


companyname= soup.find_all("div",class_="heading_6 company_name")
comp_name=[]
for i in companyname:
    for j in i.find_all("a"):
        comp_name.append(i.text.replace("\n",""))
        
All_Details=soup.find_all("div",class_="item_body")

details=[]
for i in All_Details:
    details.append(i.text.replace("\n",""))
ctc=details[1:len(details):3]
apply_date=details[2:len(details):3]
apply_date

import pandas as pd

JOB=pd.DataFrame({})
JOB['Job title']=j_title
JOB['Company Name']=comp_name
JOB['CTC']=ctc
JOB['Apply Date']=apply_date
JOB
```
